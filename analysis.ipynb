# Analysis of Fingers  

This Christmas I found myself at a friend's house party getting drunk to the old French game of [fingers](https://en.wikipedia.org/wiki/Fingers_(game)). The rules of the game are simple: each round, we all simultaenously raise either zero or five fingers, and the player whose turn it is tries to guess the *total number of fingers* raised. If the player guesses the number correctly, they win that round and leave the game. The last player remaining is the loser and, this game being invented by the French, downs a glass of wine. Well, I was quite a few eggnog shots in when my friend explained this game to me and asked me to start it off. Scared by the idea of another shot, I was reluctant to join, until he pressed me by saying "relax, the first person never loses." To me, it seemed that the first player didn't necessarily have the probabilistic advantage of winning the game. My reasoning was that, if a player were smart enough, on their turn of guessing they would withhold from raising any fingers, so the expected (and most probable) value for the sum of raised fingers would be $\sum _{i=0} ^ {n - 1} \frac{0 + 5}{2}$. Now, that is not a terribly low value in itself. Furthermore, if the player in front of you wins and leaves the game, the probability of you guessing correctly climbs even higher. So, would it be terribly unreasonable to assume that by the time we finish the first round of guesses, a good number of players would have exited the game, leaving me, the first player, in a position vulnerable to lose?  
My French friend, being drunk and French, stubbornly refused to see my point and argued the contrary: I, as the first player, will ALWAYS have the probabilistic advantage. I couldn't figure out the math to refute his response in my drunken stupor then so hence we find ourselves in this notebook.
